```python
train['연도'] = train['시점'].apply(lambda x : x[:4])
train['월'] = train['시점'].apply(lambda x : x[4:6])
train['순'] = train['시점'].apply(lambda x : x[6:])
train['품목_품종_등급'] = train.apply(lambda x : x['품목명'] + '_' + x['품종명'] + '_' + x['등급'], axis=1)
```

```python
for item in train['품목명'].unique():
    print('----', item, '----')
    print(train[train['품목명'] == item].groupby(['품종명', '등급'])['시점'].nunique())
```

```python
len_144 = []
max_len = 0
for item_ in train['품목_품종_등급'].unique():
    length = len(train[train['품목_품종_등급'] == item_])
    if length > 144:
        len_144.append(item_)
        if length > max_len:
            max_len = length
            max_item = item_
```

```python
for item_ in len_144:
    print(item_)
    print(len(train[train['품목_품종_등급'] == item_]))
    print(train[train['품목_품종_등급'] == item_]['거래단위'].nunique())
    print('--------------')    
```

```python
for item_ in train['품목_품종_등급'].unique():
    if item_ not in len_144:
        print(item_)
        print(len(train[train['품목_품종_등급'] == item_]))
        print(train[train['품목_품종_등급'] == item_]['거래단위'].nunique())
        print('--------------')    
```

```python
len_144_down = []
max_len = 0
for item_ in train['품목_품종_등급'].unique():
    length = len(train[train['품목_품종_등급'] == item_])
    if length < 144:
        len_144_down.append(item_)
```

```python
train['품목_품종_등급_단위'] = train.apply(lambda x : x['품목_품종_등급'] + '_' + x['거래단위'], axis=1)
train.head()
```

```python
unit_not_0 = []
for i in train.품목_품종_등급_단위.unique():
    if train[train['품목_품종_등급_단위'] == i]['거래단위'].nunique() != 1:
        unit_not_0.append(i)
        print(i)
        print('---------------')
print('없음')
```

```python
len_not_144 = []
for i in train.품목_품종_등급_단위.unique():
    if train[train['품목_품종_등급_단위'] == i]['시점'].nunique() != 144:
        len_not_144.append(i)
        print(i)
        print('---------------')
print('없음')
```

```python
train = train[['연도', '월', '순', '품목_품종_등급_단위', '평년 평균가격(원)', '평균가격(원)']]
train.head()
```

```python
def transform_day(x):
    if x == '상순':
        return '01'
    elif x == '중순':
        return '11'
    else:
        return '21'
```

```python
train['순'] = train['순'].apply(lambda x : transform_day(x))
train.head()
```

```python
train['시점'] = train['연도'] + '_' + train['월'] + '_' + train['순']
train
```

```python
train.drop(columns=['품목_품종_등급'], inplace=True)
train.head()
```

```python
train.to_csv('preprocessing/train_preprocessing_1.csv', index=False)
```
